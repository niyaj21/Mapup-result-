```python

def calculate_distance_matrix(csv_file):
    # Read the CSV file into a DataFrame
    data = pd.read_csv('dataset-3.csv')

    # Create a dictionary to store distances between toll locations
    distances = {}

    # Iterate through the dataset and populate the distances dictionary
    for index, row in data.iterrows():
        # Extract data from the row
        start_id = row['id_start']
        end_id = row['id_end']
        distance = row['distance']

        # Add the distance between start and end IDs
        distances[(start_id, end_id)] = distance
        distances[(end_id, start_id)] = distance  # Accounting for bidirectional distances

    # Find all unique IDs
    unique_ids = sorted(set(data['id_start']).union(set(data['id_end'])))

    # Create an empty DataFrame for distance matrix
    distance_matrix = pd.DataFrame(0, index=unique_ids, columns=unique_ids)

    # Populate the distance matrix with cumulative distances along known routes
    for i in unique_ids:
        for j in unique_ids:
            if i != j:
                distance_matrix.loc[i, j] = distances.get((i, j), 0) + distances.get((j, i), 0)

    return distance_matrix

#output
resulting_matrix = calculate_distance_matrix('dataset-3.csv')
print(resulting_matrix)
```

```python

def unroll_distance_matrix(distance_matrix):
    # Get the index of the distance matrix (contains unique IDs)
    unique_ids = distance_matrix.index

    # Create an empty list to store unrolled distances
    unrolled_distances = []

    # Iterate through unique IDs to generate combinations and distances
    for i in unique_ids:
        for j in unique_ids:
            if i != j:
                distance = distance_matrix.loc[i, j]
                unrolled_distances.append({'id_start': i, 'id_end': j, 'distance': distance})

    # Create a DataFrame from the unrolled_distances list
    unrolled_df = pd.DataFrame(unrolled_distances)

    return unrolled_df
#output
unrolled_result = unroll_distance_matrix(resulting_matrix)
print(unrolled_result)
```
